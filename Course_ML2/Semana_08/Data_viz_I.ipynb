## La 🔔 Normal

### Mean $\frac{\sum_{i=i}^{n} x_{i}}{n}$

### Std $\sqrt\frac{\sum_{i=i}^{n} (x_{i} - \overline{x})^2}{n}$

```python
pd.Series(np.random.normal(0, 1, 100000)).plot(kind='kde');
```

```python
pd.Series(np.random.normal(0, 5, 100000)).plot(kind='kde');
```

```python
h_mean = np.mean(df.heigh)
h_std = np.std(df.heigh)
h_mean, h_std
```

```python
h_mean = np.mean(df.heigh[df.sex == ' Male'])
h_std = np.std(df.heigh[df.sex == ' Male'])
h_mean, h_std
```

```python
h_mean = np.mean(df.heigh[df.sex == ' Female'])
h_std = np.std(df.heigh[df.sex == ' Female'])
h_mean, h_std
```

```python
def sample_heigh(sex):
    if sex == ' Male':
        return np.random.normal(70, 3)
    else:
        return np.random.normal(64.5, 2.5)
```

```python
df['heigh'] = df.sex.map(sample_heigh)
df
```
