## La 🔔 Normal

La distribucion Normal presenta las siguientes caracteristicas:

- Es simetrica

- Es centrada respecto a un valor

- El 95% de los datos se encuentra a 2 veces la desviacion standar de distancia de la media.

La media es la suma de todos los valores de una variable dividido por la cantidad de valores.

##### Mean $\frac{\sum_{i=i}^{n} x_{i}}{n}$

La desviacion estandar nos indica que tan separados estan los datos respecto de la media. Se calcula como sigue: es la diferencia de cada muestra a la media elevada al cuadrado, luego se suman todas las diferencias, se divide por la cantidad de muestras y finalmente se obtiene la raiz cuadrada.

##### Std $\sqrt\frac{\sum_{i=i}^{n} (x_{i} - \overline{x})^2}{n}$

```python
# distribucion de la Normal Estandard
pd.Series(np.random.normal(0, 1, 100000)).plot(kind='kde');
```

```python
# distibucion normal con una desviacion estandar mas amplia
pd.Series(np.random.normal(0, 5, 100000)).plot(kind='kde');
```

__1. Calcular la media y la std de heigh__

```python
h_mean = np.mean(df.heigh)
h_std = np.std(df.heigh)
h_mean, h_std
```

```python
h_mean = np.mean(df.heigh[df.sex == ' Male'])
h_std = np.std(df.heigh[df.sex == ' Male'])
h_mean, h_std
```

```python
h_mean = np.mean(df.heigh[df.sex == ' Female'])
h_std = np.std(df.heigh[df.sex == ' Female'])
h_mean, h_std
```

```python
def sample_heigh(sex):
    if sex == ' Male':
        return np.random.normal(70, 3)
    else:
        return np.random.normal(64.5, 2.5)
```

```python
df['heigh'] = df.sex.map(sample_heigh)
df
```
