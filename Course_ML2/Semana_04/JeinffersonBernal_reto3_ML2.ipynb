## Objetivo

## Tarea 1

Pasar valores de price que estan como cadenas de texto a números.

```python
# librerias
import pandas as pd
import numpy as np
```

```python
# cargar datos
df = pd.read_excel('./audible.xlsx', index_col=0)
df.sample(5)
```

```python
# mostrar el tipo de dato para cada columna
df.info()
```

```python
# listar registros que contienen como valor 'Free' en price
df[df['price'] == 'Free'].index.tolist()
```

```python
# convertir registros con valor 'Free' en 0
df.loc[df['price'] == 'Free', 'price'] = 0
df[df['price'] == 'Free'].index.tolist()
```

```python
# verificar los tipos de datos que hay en Price
df['price'].apply(type).value_counts()
```
