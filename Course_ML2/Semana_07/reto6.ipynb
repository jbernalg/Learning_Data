En este conjunto de datos encontrará una colección completa de información sobre más de 4700 propiedades residenciales disponibles, que incluyen casas, apartamentos y pisos ofrecidos en alquiler en India. Esto refleja la importante expansión experimentada en el sector de la vivienda de la India, que ha ido en paralelo a la trayectoria ascendente de los niveles de ingresos en el país. El conjunto de datos es rico en diversos atributos, como el número de dormitorios (BHK), precios de alquiler, tamaño de la propiedad, número de plantas, tipo de área, localidad, ciudad, estado del mobiliario, preferencias del inquilino, número de baños e información de contacto del punto de contacto correspondiente.

- BHK: Número de dormitorios, salón y cocina. 

- Rent: Alquiler de las casas/apartamentos/pisos. 

- Size: Tamaño de las casas/apartamentos/pisos en pies cuadrados. 

- Floor: Planta en la que se encuentran las casas/apartamentos/pisos y número total de plantas (Ejemplo: planta baja de 2, planta 3 de 5, etc.). 

- Area Type: Tamaño de las casas/apartamentos/pisos calculado en función de la superficie total, la superficie útil o la superficie construida. 

- Area Locality: Localidad de las casas/apartamentos/pisos. 

- City: Ciudad donde se encuentran las casas/apartamentos/pisos.

- Furnishing Status: Estado del mobiliario de las casas/apartamentos/pisos: amueblado, semi-amueblado o sin amueblar. 

- Tenant Preferred: Tipo de inquilino preferido por el propietario o agente. 

- Bathroom: Número de baños. 

- Point of Contact: Con quién debe ponerse en contacto para obtener más información sobre las casas/apartamentos/pisos.
