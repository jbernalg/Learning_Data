Lee el archivo 𝐺𝐿𝑂𝐵𝐴𝑁𝑇.𝑐𝑠𝑣 línea por línea, y a medida que vayas leyendo, escribe en 𝑎𝑛𝑎𝑙𝑖𝑠𝑖𝑠_𝑎𝑟𝑐ℎ𝑖𝑣𝑜.𝑐𝑠𝑣 lo siguiente:

    - ESTABLE: Si 𝐶𝑙𝑜𝑠𝑒𝑖 −𝑂𝑝𝑒𝑛𝑖 = 0 (donde 𝑖 es el día 𝑖 que se está analizando).

```python
# libreria
import csv, json

# leer GLOBANT.csv y crear analisis.csv
with open('GLOBANT.csv', 'r') as globant, open('analisis.csv', 'w') as analisis:
    
    # leer archivo globant
    reader = csv.reader(globant)
    # saltar encabezado
    next(reader)

    # agregar encabezado al archivo nuevo
    analisis.write('Fecha\tComportamiento de la accion\tPunto medio HIGH-LOW\n')
    # lista para almacenar datos para futuros calculos
    rows = []

    # iterar sobre cada fila de GLOBANT
    for fila in reader:
        # omitir fechas o filas vacias
        if not fila or fila[0].strip() == '':
            continue
        # extraer los valores requeridos
        date = fila[0]
        open_ = float(fila[1])
        high = float(fila[2])
        low = float(fila[3])
        close = float(fila[4])

        # calcular diferencia entre el precio de cierre y el de apertura
        diferencia = close - open_
        # clasificar comportamiento
        if diferencia > 0:
            comportamiento = 'SUBE'
        elif diferencia < 0:
            comportamiento = 'BAJA'
        else:
            comportamiento = 'ESTABLE'
        
        # calcular el punto medio entre el precio maximo y minimo
        punto_medio = (high - low) / 2

        # escribir resultados en analisis
        analisis.write(f'{date}\t{comportamiento}\t{punto_medio}\n')
        # guardar informacion relevante en rows
        rows.append((date, high, low, comportamiento))

        
```

```python
import csv
import json

def solucion():
    # --- Leer GLOBANT.csv y crear analisis_archivo.csv ---
    with open('GLOBANT.csv', 'r') as globant, \
         open('analisis_archivo.csv', 'w') as analisis:
        
        reader = csv.reader(globant)
        next(reader)  # Saltar encabezado original
        
        # Escribir encabezado nuevo
        analisis.write("Fecha\tComportamiento de la accion\tPunto medio HIGH-LOW\n")
        
        rows = []
        for row in reader:
            if not row or row[0].strip() == '':
                continue
            date  = row[0]
            open_ = float(row[1])
            high  = float(row[2])
            low   = float(row[3])
            close = float(row[4])
            
            diff = close - open_
            if diff > 0:
                comportamiento = "SUBE"
            elif diff < 0:
                comportamiento = "BAJA"
            else:
                comportamiento = "ESTABLE"
            
            punto_medio = (high - low) / 2
            
            analisis.write(f"{date}\t{comportamiento}\t{punto_medio}\n")
            rows.append((date, high, low, comportamiento))
    
    # --- Calcular valores para detalles.json ---
    lowest_price  = min(rows, key=lambda x: x[2])   # por Low
    highest_price = max(rows, key=lambda x: x[1])   # por High
    
    cant_sube    = sum(1 for r in rows if r[3] == "SUBE")
    cant_baja    = sum(1 for r in rows if r[3] == "BAJA")
    cant_estable = sum(1 for r in rows if r[3] == "ESTABLE")
    
    detalles = {
        "date_lowest_price":    lowest_price[0],
        "lowest_price":         lowest_price[2],
        "date_highest_price":   highest_price[0],
        "highest_price":        highest_price[1],
        "cantidad_veces_sube":  cant_sube,
        "cantidad_veces_baja":  cant_baja,
        "cantidad_veces_estable": cant_estable
    }
    
    with open('detalles.json', 'w') as f:
        json.dump(detalles, f, indent=4)

solucion()
```
