# __Reto 5: Adicion de variables__

## Objetivo

Saber qué método de codificación usar cuando se enfrenta a variables categóricas nominales y ordinales 

## Tarea 1

Cargar el dataset modificado Breast Cancer

```python
# librerias
import pandas as pd
import numpy as np
```

```python
# cargar datos
df = pd.read_excel('./breast-cancer-modified.xlsx', index_col=0)
df.sample(5)
```

## Tarea 2

Aplicar la codificacion correspondiente dependiendo de la informacion de cada variable

__age: Ordinal__

```python
df.age.unique()
```

```python
# importar OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

# instanciar el objeto ordinalencoder
# agregar en orden las categorias
encoder = OrdinalEncoder(categories=[["20-29'", "30-39'", "40-49'", "50-59'", "60-69'",  "70-79'"]])

# ajustar y transformar el encoder
age_ordinal = encoder.fit_transform(df[['age']])

# concatenar valores codificados el dataframe original
df['age_encoded'] = age_ordinal

df.sample(5)
```

__Variables Nominales: menopause, breast, irradiat, Class__

```python
# codificar todas las variables nominales en un mismo paso
var_nominal = pd.get_dummies(df[['menopause', 'breast', 'irradiat', 'Class']])
var_nominal
```

```python
# concatenar variables codificadas al dataframe original
df = pd.concat([df, var_nominal], axis=1)
df.sample(5)
```

__deg-malig: Ordinal__

```python
# obtener valores unicos de la variable deg-malig
df['deg-malig'].unique()
```

```python
# instanciar el objeto ordinalencoder
# agregar en orden las categorias
encoder2 = OrdinalEncoder(categories=[["1'", "2'", "3'"]])

# ajustar y transformar el encoder
deg_ordinal = encoder2.fit_transform(df[['deg-malig']])

# concatenar valores codificados el dataframe original
df['deg_malig_encoded'] = deg_ordinal

df.sample(5)
```
