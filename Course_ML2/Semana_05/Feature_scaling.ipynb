# __Transformación de datos (Feature scaling)__

Vamos a implementar MinMaxScaler y StandardScaler de Sklearn, tanto en un dataset de juguete como en el dataset `titanic.csv`

```python
# Ubicación de los dataset
path_juguete = "https://drive.google.com/uc?id=1pfwAP_lYaS2EgoZGUtjzmy35QBrynQ6I&export=download"
path_titanic = "https://drive.google.com/uc?id=1KsLlUKjtnin2N2_EHugykz-wBaP8Gh4L&export=download"
```

## Lectura de los dataset

```python
# pandas nos permite leer los dataset fácilmente
import pandas as pd
```

```python
# Leer ambos dataset de una vez
df_juguete = pd.read_csv(path_juguete)
df_titanic = pd.read_csv(path_titanic)
```

```python
# mostrar dataset de juguete
df_juguete
```

```python
# mostrar dataset del titanic
df_titanic.sample(5)
```

## ¡A transformar!

Vamos a usar la librería de sklearn para realizar las transformaciones (Min - Max Scaler y Standard Scaler) en el dataset de juguete.

```python
# importar transformaciones
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

En el siguiente paso, instanciamos dos objetos de cada clase importada

```python
# Instanciación de los objetos escaladores
scaler_minmax = MinMaxScaler()
scaler_standard = StandardScaler()
```

## Standard Scaler

El primer método que vamos a ver es `fit_transform` de `StandardScaler` (Si aprendes a usar este, `MinMaxScaler` será pan comido porque ambas clases se usan prácticamente igual).
Como es de costumbre, es importante que leas la documentación, esto te permitirá comprender qué tanto puedes hacer con estas clases.

A continuación te dejamos las dos documentaciones:


*   [Documentación de MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html#:~:text=Transform%20features%20by%20scaling%20each,e.g.%20between%20zero%20and%20one.)
*   [Documentación de StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)




```python
# Transformamos el dataset
df_scaled = scaler_standard.fit_transform(df_juguete)

# Mostramos el dataset transformado
df_scaled
```

Si quisiéramos conocer las varianzas de cada columna antes de ser transformada, podemos conocerlas imprimiendo el atributo `var_`

**Nota: La desviación estándar se define como la raíz cuadrada de la varianza, es decir:** $\sigma=\sqrt{var}$

```python
# Imprimimos las varianzas
scaler_standard.var_
```

```python
# Importamos numpy para aplicar operaciones vectoriales (Sacar la raíz de un vector)
import numpy as np

# Sacamos la raíz de cada componente
np.sqrt(scaler_standard.var_)
```

Ahora, para conocer las medias de cada columna antes de ser escalada, imprimimos el atributo `mean_`

```python
# Imprimimos las medias
scaler_standard.mean_
```

Cabe resaltar que el método `fit_transform` se olvida de los nombres de las columnas (`x1` y `x2` en este caso) porque el resultado que retorna es un objeto de la clase `numpy.ndarray`, ¡Vamos a comprobarlo en la siguiente celda de código!

```python
# Imprimimos el tipo de dato que guarda la variable df_scaled
print( type(df_scaled) )
```

Para volver a colocarle nombre a las columnas, hacemos lo siguiente

```python
df_scaled = pd.DataFrame(df_scaled, columns=df_juguete.columns)
df_scaled
```

## MinMaxScaler

```python
# transformamos el dataset con minmaxscaler
df_scaled_minmax = scaler_minmax.fit_transform(df_juguete)
df_scaled_minmax
```

Para ponerle nuevamente el nombre a las columnas ejecutamos la siguiente celda de código

```python
df_scaled_minmax = pd.DataFrame(df_scaled_minmax, columns=df_juguete.columns)
df_scaled_minmax
```
