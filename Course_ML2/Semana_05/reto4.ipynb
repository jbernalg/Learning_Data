## Objetivo

Aplicar las dos técnicas de escalado de datos vistas en la semana 5

## Tarea 1

Cargar el dataset del titanic y eliminar las columnas 'Name' y 'Sex'.

```python
# librerias
import pandas as pd
import numpy as np
```

```python
# cargar datos
df = pd.read_csv('./titanic.csv')
df.sample(5)
```

```python
# eliminar variables Name y Sex
df.drop(columns=['Name', 'Sex'], inplace=True)
df.sample(5)
```

## Tarea 2

Aplicar MinMaxScaler sobre el dataframe resultante de eliminar las dos columnas del punto 1 (no olvidar volver a ponerle el nombre a las columnas)

```python
# importar transformaciones
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Instanciar los objetos escaladores
scaler_minmax = MinMaxScaler()
scaler_standard = StandardScaler()
```

```python
# transformar el dataset con minmaxscaler
df_scaled_minmax = scaler_minmax.fit_transform(df)
df_scaled_minmax
```

```python
# agregar el nombre de las columnas
df_scaled_minmax = pd.DataFrame(df_scaled_minmax, columns=df.columns)
df_scaled_minmax
```

## Tarea 3

Aplicar StandardScaler sobre el dataframe resultante de eliminar las dos columnas del punto 1 (no olvidar volver a ponerle el nombre a las columnas). 

```python
# transformar el dataset con Standard Scaler
df_scaled = scaler_standard.fit_transform(df)
df_scaled
```

```python
# agregar los nombres a las columnas
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled
```

Extraer la desviación estándar y la media aritmética de cada columna (usando los métodos y atributos proporcionados por la clase StandardScaler y numpy).
